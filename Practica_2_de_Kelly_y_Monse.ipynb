```python
#Nuestros nombres son:
#Kelly Santos García y Tania Monserrat Gopar Iniesta
#Aquí esta todo lo que importamos
import random

#Ahora si empieza nuestra practica
class Cliente:
    def __init__(self, id_cliente, premier=False):
        self.id_cliente = id_cliente
        self.premier = premier
        self.tiempo_llegada = random.uniform(1, 3)  # Tiempo interarribo en minutos
        self.tiempo_atencion = random.uniform(1, 81)  # Tiempo de atención en minutos
        self.tiempo_espera = 0  # Tiempo que el cliente espera para ser atendido

#Aqui va la clase cola requerida
class Cola:
    def __init__(self, lineas_premier, lineas_regulares):
        self.lineas_premier = [None] * lineas_premier  # Líneas exclusivas para premier
        self.lineas_regulares = [None] * lineas_regulares  # Líneas para clientes regulares
        self.cola_premier = []
        self.cola_normal = []
        self.tiempo_actual = 0  # Reloj de simulación
        self.usuarios_atendidos = 0
        self.usuarios_premier_atendidos = 0
        self.usuarios_regulares_atendidos = 0
        self.tiempo_max_espera_premier = 0
        self.total_tiempo_atencion_premier = 0
        self.total_clientes_premier = 0

    def agregar_cliente(self, cliente):
        if cliente.premier:
            self.cola_premier.append(cliente)
        else:
            self.cola_normal.append(cliente)

    def avanzar_tiempo(self):
        # Avanza el tiempo de la simulación y procesa clientes en las líneas
        self.tiempo_actual += 1

        # Procesar líneas premier
        for i in range(len(self.lineas_premier)):
            if self.lineas_premier[i] is not None:  # Si la línea está ocupada
                cliente = self.lineas_premier[i]
                cliente.tiempo_atencion -= 1  # Reducir el tiempo de atención restante

                # Si terminó de ser atendido
                if cliente.tiempo_atencion <= 0:
                    self.lineas_premier[i] = None  # Liberar la línea
                    self.usuarios_premier_atendidos += 1
                    self.total_tiempo_atencion_premier += cliente.tiempo_espera
                    self.total_clientes_premier += 1
                    self.tiempo_max_espera_premier = max(self.tiempo_max_espera_premier, cliente.tiempo_espera)
                    self.usuarios_atendidos += 1

        # Procesar líneas regulares
        for i in range(len(self.lineas_regulares)):
            if self.lineas_regulares[i] is not None:  # Si la línea está ocupada
                cliente = self.lineas_regulares[i]
                cliente.tiempo_atencion -= 1  # Reducir el tiempo de atención restante

                # Si terminó de ser atendido
                if cliente.tiempo_atencion <= 0:
                    self.lineas_regulares[i] = None  # Liberar la línea
                    self.usuarios_regulares_atendidos += 1
                    self.usuarios_atendidos += 1

    def procesar_clientes(self):
        # Asignar líneas a clientes premier
        for cliente in self.cola_premier[:]:
            for i in range(len(self.lineas_premier)):
                if self.lineas_premier[i] is None:  # Asignar a una línea vacía
                    cliente.tiempo_espera = self.tiempo_actual - cliente.tiempo_llegada
                    self.lineas_premier[i] = cliente
                    self.cola_premier.remove(cliente)
                    break

        # Asignar líneas a clientes regulares solo si no hay premier en espera
        if not self.cola_premier:
            for cliente in self.cola_normal[:]:
                for i in range(len(self.lineas_regulares)):
                    if self.lineas_regulares[i] is None:  # Asignar a una línea vacía
                        cliente.tiempo_espera = self.tiempo_actual - cliente.tiempo_llegada
                        self.lineas_regulares[i] = cliente
                        self.cola_normal.remove(cliente)
                        break

    def imprimir_resultados(self):
        tiempo_promedio_premier = self.total_tiempo_atencion_premier / self.total_clientes_premier if self.total_clientes_premier > 0 else 0
        print("\nResultados de la simulación:")
        print(f"Usuarios atendidos: {self.usuarios_atendidos}")
        print(f"Usuarios premier atendidos: {self.usuarios_premier_atendidos}")
        print(f"Usuarios regulares atendidos: {self.usuarios_regulares_atendidos}")
        print(f"Tiempo máximo de espera (premier): {self.tiempo_max_espera_premier:.2f} minutos")
        print(f"Tiempo promedio de atención (premier): {tiempo_promedio_premier:.2f} minutos")

def simulacion(num_clientes):
    # Definir el número de líneas para premier y regulares antes de empezar
    lineas_premier = 90  # Número fijo de líneas para premier
    lineas_regulares = 20  # Número fijo de líneas para regulares

    print(f"Número de clientes: {num_clientes}")
    print(f"Líneas premier: {lineas_premier}, Líneas regulares: {lineas_regulares}")

    centro = Cola(lineas_premier, lineas_regulares)

    # Agregar todos los clientes a las colas
    for i in range(1, num_clientes + 1):
        premier = True if random.randint(1, 6) == 1 else False
        cliente = Cliente(i, premier=premier)
        centro.agregar_cliente(cliente)

    # Simulación a lo largo del tiempo
    while centro.cola_premier or centro.cola_normal or any(centro.lineas_premier) or any(centro.lineas_regulares):
        centro.avanzar_tiempo()  # Avanza un minuto en la simulación
        centro.procesar_clientes()  # Intentar atender a los clientes

    # Imprimir los resultados al final
    centro.imprimir_resultados()

# Ejecución de la simulación con un número aleatorio de clientes entre 500 y 1000
if __name__ == "__main__":
    num_clientes = random.randint(500, 1000)  # Número de clientes aleatorio entre 500 y 1000
    simulacion(num_clientes)
```
